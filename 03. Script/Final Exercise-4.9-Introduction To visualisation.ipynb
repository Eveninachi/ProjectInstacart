##Combine df_customers with orders_products_merged_updated_2.pkl

```python
df_ords_prods_merged = pd.read_pickle(os.path.join(path,'02. Data\Prepared data\orders_products_merged_updated.pkl'))
```

```python
df_ords_prods_merged.info()
```

```python
#Change the data type of user_id to string
df_ords_prods_merged['user_id'] = df_ords_prods_merged['user_id'].astype('str')
```

```python
#Change the data type of order_id to string
df_ords_prods_merged['order_id'] = df_ords_prods_merged['order_id'].astype('str')
```

```python
df_ords_prods_merged.info()
```

```python
#Drop the _merge column from the df_ords_prods_merge dataframe as it is not needed. 
df_ords_prods_merged = df_ords_prods_merged.drop(columns = ['_merge'])
```

```python
df_ords_prods_merged.info()
```

```python
#Merge df_ords_prods_merge with df_customers
df_ords_prods_all = df_ords_prods_merged.merge(df_customers, on=['user_id'], indicator=True)
```

```python
#Check the frequency of the new dataframe
df_ords_prods_all['_merge'].value_counts()
```
